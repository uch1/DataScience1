```python
# Read Titanic Data Set and create a dataframe using csv 
titanic_cvs = pd.read_csv('downloads/titanic.csv')
titanic_df = pd.DataFrame(data=titanic_cvs)
```

```python
def probability(event_count, sample_space):
    return event_count/sample_space

def percentage(prob):
    percentage = prob * 100
    return '{}% chance of occurence'.format(percentage)

def fraction_probability(event_count,sample_space):
    denominator = round( sample_space/event_count)
    numerator = int(event_count/event_count)
    return '{} /{} chance of occurence'.format(numerator, denominator)
```

```python
# 2. Given that someone is on the top deck, what are the chances they survived?
```

```python
# 3. Given that someone has a sibling or parent, what are the chances they survived?
passenger_with_SibPar_df = titanic_df[(titanic_df['SibSp'] > 0) & (titanic_df['Parch'] > 0)]
passenger_with_SibPar_count = len(passenger_with_SibPar_df)
passenger_with_SibPar_count
```

```python
passenger_with_SibPar_df
```

```python
survived_passenger_with_sibpar_df = passenger_with_SibPar_df[passenger_with_SibPar_df['Survived'] == 1]
survival_count = len(survived_passenger_with_sibpar_df)
survival_count
```

```python
display(percentage(probability(survival_count,passenger_with_SibPar_count)))
```

```python
# 4. Given that someone is older than 18, what are the chances they survived?
above_age18_count = len(titanic_df[titanic_df['Age'] >= 18])
display(above_age18_count)
```

```python
survived_above_age18_count = len(titanic_df[(titanic_df['Age'] >= 18) & (titanic_df['Survived'] == 1)])
display(survived_above_age18_count)
```

```python
display(percentage(probability(survived_above_age18_count,above_age18_count)))
```

```python
# 5. Given all the passengers, what is the probability that the passenger was a male and survived? 
# Get every male passengers on the ship 
male_passengers_df = titanic_df[titanic_df['Sex'] == 'male']
# Get the total count of male passengers
male_passengers_count = len(male_passengers_df)
display(male_passengers_count)
```

```python
# Find and get the total count of male survivors 
male_survivors_df = male_passengers_df[male_passengers_df['Survived'] == 1]
male_survivors_count = len(male_survivors_df)
display(male_survivors_count)
```

```python
# what is the probability that the passenger was a male and survived?
male_survivor_prob = probability(male_survivors_count,male_passengers_count)
display(percentage(male_survivor_prob))
```

```python
# 6. Given a passenger embarked on SouthHampton, what is the probability that he/she is young <12

# Find passengers who embarked on SouthHampton 
southhampton_passengers_df = titanic_df[titanic_df['Embarked'] == 'S']
southhampton_passengers_df
```

### Find passengers whose ages were below 12 years

### Male Passengers under the age of 12

### Female Passengers under the age of 12 

```python

female_passengers_under_12_df = passengers_under_12[passengers_under_12['Sex']=='female']
female_passengers_under_12_count = len(female_passengers_under_12_df)
female_passengers_under_12_count
```

### what is the probability that she is young <12

```python

female_prob = probability(female_passengers_under_12_count, passengers_under_12_count)
display(percentage(female_prob))
```
